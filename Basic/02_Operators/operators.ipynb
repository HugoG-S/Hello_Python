### Operadores
Los operadores son símbolos que leindican al intérprete que realice una operación específica.

```python
### Operadores Aritmeticos. ###
## Un operadore aritmético toma dos operandos como entrada, realiza un cálculo y devuelve el resultado

print(3 + 4)
print(3 - 4)
print(3 * 4)
print(3 / 4)
print(10 % 2)
print(10 % 3)
```

```python
# Realiza la división y termina aproximada a un número entero
print(10 // 3) 
```

```python
# Calcular un exponente
print(2 ** 3) 

print(2 ** 3 + 3 - 7 / 1 //4)
```

```python
# Concatenamos los string
print("Hola " + "Python") 
```

```python
# Convertirmos el int 5 a string para poder concatenarlos
print("Hola " + str(5)) 

print("Hola " * 5)
print("Hola " * (2 ** 3))
```

```python
# Utilizamos int() para transformar el valor de my_float de float a int
my_float = 2.5 * 2
print("Hola " * int(my_float))
```

```python
### Operadores Comparativos. ###
## Un operador comparativo o relacional se emplea para comprar y establecer la realción entre ellos. 
# Devuelve un valor booleano (true o false) basado en la condición

print(3 > 4)    # Mayor que
print(3 < 4)    # Menor que
print(3 >= 4)   # Mayor o igual
print(3 <= 4)   # Menor o igual
print(3 == 4)   # Igual que
print(3 != 4)   # Diferente de
```

```python
# También se puede usar varios operadores comparativos
print(3 > 4 == 2)
```

```python
# Podemos usarlos también para comparar entre string
print("Hola" > "Python")
print("Hola" < "Python")
print("Hola" >= "Python")
print("Hola" <= "Python")
print("Hola" == "Python")
print("Hola" != "Python")

print("Hola" >= "Zola") # Ordenación alfabética
print(len("Hola") >= len("Zola")) # Ordenación caracteres
```

```python
### Operadores Lógicos ###
## Se utiliza un operador lógico para tomar una decisión basada en mútiples condiciones. Los operadores 
# lógicos utilizados en Python son and, or y not.

print(3 > 4 and "Hola" > "Python") 
print(3 > 4 or "Hola" > "Python")
print(3 < 4 and "Hola" < "Python") 
print(3 < 4 or "Hola" > "Python")
print(3 < 4 or "Hola" > "Python" and 4 == 4)

# Not niega la condición
print(not(3 > 4))
```
